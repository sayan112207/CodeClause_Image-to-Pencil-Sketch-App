# Importing Libraries

```python
# Importing the required libraries
import cv2
import matplotlib.pyplot as plt
```

# Loading the image

```python
# Loading the input image
img = cv2.imread('image.jpg')

# Convert from BGR to RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img)
plt.axis('off')
plt.show()
```

# Converting to GrayScale

```python
# Converting the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
plt.axis('off')
plt.show()
```

# Inverting the image

```python
# Inverting the grayscale image
inverted_gray_img = 255 - gray_img
plt.imshow(inverted_gray_img)
plt.axis('off')
plt.show()
```

# Applying Gaussian blur to the inverted image

```python
blurred_img = cv2.GaussianBlur(inverted_gray_img, (21, 21), 0)
plt.imshow(blurred_img)
plt.axis('off')
plt.show()
```

# Converting the blurred image to pencil sketch using the divide blend mode

```python
pencil_sketch_img = cv2.divide(gray_img, 255 - blurred_img, scale=256)
plt.imshow(pencil_sketch_img)
plt.axis('off')
plt.show()
```

# Displaying the input and the final image

```python
# Displaying the input and output images

#The Input Image
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Input Image')
plt.axis('off')

#The Pencil Sketch
plt.subplot(1,2,2)
plt.imshow(pencil_sketch_img, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')
plt.show()

# Saving the output image
cv2.imwrite('output_image.jpg', pencil_sketch_img)
```
